```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
import os
```

```python
os.chdir(r"C:\python\first project")
```

```python
data=pd.read_csv('salary_dataset.csv')
```

```python
sns.pairplot(data)
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,1]
x_train,x_test ,y_train,y_test =train_test_split(x,y, train_size=0.8,random_state=10)
```

```python
my_model=LinearRegression()
my_model.fit(x_train,y_train)
my_model.score(x_train,y_train)
my_model.score(x_test,y_test)
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,my_model.predict(x_train),color='black')
plt.title("First Project")
plt.xlabel("Years Of Expeience")
plt.ylabel("Salary")
```

```python
y_pred=my_model.predict(x_test)
plt.scatter(x_test,y_test,color='black')
plt.plot(x_test,y_pred,color='red')
plt.title("My Company")
plt.xlabel("Year Of Experience")
plt.ylabel("Salary")
```

```python
c=[i for i in range(1,len(y_test)+1)]
plt.plot(c,y_test,color="black",linestyle='-')
plt.plot(c,y_pred,color="red",linestyle='-')
plt.title("Difference")
plt.xlabel("index")
plt.ylabel("Salary")
```

```python
c=[i for i in range(1,len(y_test)+1)]
plt.plot(c,y_test-y_pred,color="yellow",linestyle="-")
plt.title("Error")
plt.xlabel("Index")
plt.ylabel("Salary")
```

```python
mse=mean_squared_error(y_test,y_pred)
r=r2_score(y_test,y_pred)
print(mse)
print(r)
```
